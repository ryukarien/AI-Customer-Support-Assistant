```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load datasets
df_train = pd.read_csv("data/processed/train.csv")
df_val = pd.read_csv("data/processed/val.csv")
df_test = pd.read_csv("data/processed/test.csv")

# Step 3: General info
print("=== Train Dataset Info ===")
df_train.info()
print("\n=== Validation Dataset Info ===")
df_val.info()
print("\n=== Test Dataset Info ===")
df_test.info()

# Step 4: Check for missing values
print("\n=== Missing Values (Train) ===")
print(df_train.isnull().sum())
print("\n=== Missing Values (Validation) ===")
print(df_val.isnull().sum())
print("\n=== Missing Values (Test) ===")
print(df_test.isnull().sum())

# Step 5: Sample few rows
print("\n=== Train Sample Messages ===")
print(df_train.sample(5))

# Step 6: Check class distribution (if you have a 'label' column)
if 'label' in df_train.columns:
    print("\n=== Class Distribution (Train) ===")
    print(df_train['label'].value_counts())
    
    # Optional: plot
    plt.figure(figsize=(6,4))
    sns.countplot(x='label', data=df_train)
    plt.title("Class Distribution - Train Set")
    plt.show()
```

```python
# Step 7: Text length analysis
df_train['text_length'] = df_train['text'].apply(len)

plt.figure(figsize=(8,4))
sns.histplot(df_train['text_length'], bins=20, kde=True)
plt.title("Distribution of Tweet Lengths - Train Set")
plt.xlabel("Number of Characters")
plt.ylabel("Count")
plt.show()
```

```python
# Step 8: Convert 'created_at' to datetime and analyze
df_train['created_at'] = pd.to_datetime(df_train['created_at'])
df_train['hour'] = df_train['created_at'].dt.hour

plt.figure(figsize=(8,4))
sns.histplot(df_train['hour'], bins=24, kde=False)
plt.title("Tweet Frequency by Hour - Train Set")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Tweets")
plt.show()
```

```python
# Step 9: Numeric correlation
numeric_cols = df_train.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(8,6))
sns.heatmap(df_train[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Numeric Features")
plt.show()
```
